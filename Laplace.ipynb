```python
import numpy as np
from scipy.optimize import minimize

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
np.random.seed (1)
```

```python
sns.set_theme(style = 'darkgrid')
```

```python
BETA = np.array([-1.5, 2.0])

input_range = np.linspace(-1, 1, 100)

X = np.vstack([np.ones (100), np.linspace(-1, 1, 100)])

X = X.T

Y = X @ BETA + np.random.normal(0, 0.3, 100)
```

```python
def Absoluta(Paremetros):

    Perda = np.sum(np.abs(Y - X @ Paremetros ))

    return Perda
```

```python
U, D, V = np.linalg.svd(X, full_matrices = False)

Gaussiana = V @ np.diag(D**(-1)) @ U.T @ Y

Laplace = minimize(Absoluta, x0 = np.ones(2))['x']
```

```python
plt.figure(figsize = (10, 5))

Grafico = sns.scatterplot(x = X[:, 1], y = Y)

sns.lineplot(x = X[:, 1], y = X @ Gaussiana, color = 'red')

Grafico.spines['right'].set_visible(False)

Grafico.spines['top'].set_visible(False)

plt.title('Gaussiana')

plt.show()
```

```python
plt.figure(figsize = (10, 5))

Grafico = sns.scatterplot(x = X[:, 1], y = Y)

sns.lineplot(x = X[:, 1], y = X @ Laplace, color = 'red')

Grafico.spines['right'].set_visible(False)

Grafico.spines['top'].set_visible(False)

plt.title('Laplace')

plt.show()
```

```python
Erro = np.linalg.norm(BETA - Gaussiana)

print(f'Erro (Gaussiano): {np.round(Erro, 4)} \n')

Erro = np.linalg.norm(BETA - Laplace)

print(f'Erro (Laplace): {np.round(Erro, 4)}')
```
